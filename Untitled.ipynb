## Explore ES DSL 

### Approach1: ES-DSL Full-Text Search (more liberal)


1. each word from multi-words query is tokenized
2. **the match score only accounts for frequency**, but does not account for the distance between tokens

e.g. 
queary text = ['experienced hikers', 'good workout', 'beautiful views', 'seven-mile round trip', 'climb']

this method treats them as tokens, 'experienced', 'hikers', 'good', 'workout', 'beautiful', 'views'... 



- use [Multiword Queries](https://www.elastic.co/guide/en/elasticsearch/guide/current/match-multi-word.html)
- advantage: not all tokens require to be present in the field of the document. any document whose the field contains at least one of the specified tokens will match the query. The more tokens that match, the more relevant the document.

e.g.

s.query("match", overview = 'experienced hikers, good workout, beautiful views, seven-mile round trip, climb')


- use Controlling Precision
- advantage: minimum_should_match parameter allows you to specify the number of terms that must match for a document to be considered relevant. 

e.g.

s.query({
    "match": {
        "overview": {
            "query": 'experienced hikers, good workout, beautiful views, seven-mile round trip, climb',
            "minimum_should_match": "75%"
         }
    }
})


### Approach2: ES-DSL Proximity Matching (more conservative)

1. each word from multi-value query is tokenized
2. **all tokens require to be present in the field of the document**
3. under the condition of "2", the match score accounts for distance between tokens

e.g. 
queary text = ['experienced hikers', 'good workout', 'beautiful views', 'seven-mile round trip', 'climb']

this method treats them as tokens, 'experienced', 'hikers', 'good', 'workout', 'beautiful', 'views'... 
and all tokens require to be present. None of overview document will include all the tokens except the query


- use [multivalue fields of match_phrase](https://www.elastic.co/guide/en/elasticsearch/guide/current/_multivalue_fields_2.html)
- drawback1: it accounts for the distance not only between 'experience' and 'hikers', but also 'hikers' and 'good'
- drawback2: all tokens of keywords require to be present in the text of overview to match, **with the same order**.

e.g.

s.query("match_phrase", overview = ['experienced hikers', 'good workout', 'beautiful views', 'seven-mile round trip', 'climb'])


- use [Mixing It Up of match_phrase](https://www.elastic.co/guide/en/elasticsearch/guide/current/slop.html#slop)
- advantage: "slop" parameter introduces a degree of flexibility into phrase matching, which tells the match_phrase query how far apart terms are allowed to be while still considering the document a match. With a high enough slop value, words can be arranged in any order. By setting a high slop value like 50 or 100, you can exclude documents in which the words are really too far apart, but give a higher score to documents in which the words are closer together.

e.g. 

s.query({
    "match_phrase": {
        "overview": {
            "query": ['experienced hikers', 'good workout', 'beautiful views', 'seven-mile round trip', 'climb'],
            "slop": 100
         }
    }
})



### Approach3: ES-DSL Combination of Full-Text Search and Proximity Matching

- use [Proximity for Relevance](https://www.elastic.co/guide/en/elasticsearch/guide/current/proximity-relevance.html): use full-text search as must query and match_phrase as should query

s.query({
    "bool": {
      "must": {
        "match": { 
          "overview": {
            "query": 'experienced hikers, good workout, beautiful views, seven-mile round trip, climb',
            "minimum_should_match": "30%"
          }
        }
      },
      "should": {
        "match_phrase": {

## Things to consider

Phrase and proximity queries are more expensive than simple match queries. Whereas a match query just has to look up terms in the inverted index, a match_phrase query has to calculate and compare the positions of multiple possibly repeated terms.

For this reason, using numeric feature based model first and then sorting by keyword-ES would be faster
